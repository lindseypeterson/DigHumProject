<b>Dataset Source:</b> The dataset for this project comes from the <i>Civil War and Reconstruction Governors of Mississippi Project</i> (CWRGM), a digital documentary edition of 22,000 letters written to the Governors of Mississippi during the Civil War and Reconstruction (1859â€“1882). Originally housed at the Mississippi Department of Archives and History, CWRGM is digitizing, transcribing, and annotating these documents with support from the National Endowment for the Humanities, the National Historical Publications and Records Commission, and the Watson-Brown Foundation. As of June 2024, the site has made 11,835 documents freely available online at cwrgm.org. 

The CWRGM collection is far less about the governors, and more about the hopes, frustrations, and fears of ordinary Americans during one of the most tumultuous times in American history. The collection features documents written from nearly every constituency imaginable during the era, including enlisted men, Unionists, veterans, women, and after April 1865, African Americans. Consequently, it is a rich repository of letters, telegrams, petitions, and rations pertaining to African American history, especially the process through which emancipation and freedom was fought for and secured.

<b>Subject Tag Creation:</b> Publishing their protocols on their Editorial Process page, CWRGM adds linked subject tags from an internally controlled vocabulary to the documentsâ€™ transcriptions to increase the discoverability of marginalized voices in the collection. For instance, the subject tag African Americans has been added to 1,443 documents to ensure that these materials are found by researchers, even though the term â€œAfrican Americanâ€ never appears verbatim in the transcriptions. While subjects can be found in the document metadata, the subject tags are linked directly to a textâ€™s transcription, and unlike the projectâ€™s metadata which faithfully replicates the Library of Congressâ€™ Authorities files, the subject tags are internally created and maintained by CWRGMâ€™s editorial staff. A subject tag is only added to a document once, and all transcriptions and subject tags go through a two stage review process conducted by trained researchers, typically undergraduate and graduate students. Subject tags also mimic the LOCâ€™s authorities files wherever possible, but editors create their own taxonomy to offer reparative terminology and to allow for nesting. For example, the documents tagged with African Americans--Enslaved People and African Americans--Historically Free and Newly Freed child tags are also found under the African Americans parent tag and establish a relationship between the status of enslaved and free, and race. This nesting feature, however, does not duplicate the entries in the data file, only in the African Americans tag results at the website.

<b>Co-occurence Data Creation:</b> The CWRGM research and editorial team add subject tags in FromThePage (FTP), a crowdsourcing transcription software created by Brumfield Labs. FTP creates co-occurrence data for each subject in the collection. The data is exported to a .csv file listing the other subjects tagged to a document also containing one of the African Americans tags; it also includes identifying information about the document that was created by the CWRGM research team from transcriptions and metadata. To create the dataset for this project, I downloaded and collated the co-occurrences files for four subject tags pertaining to African Americans under the Social Identifiers category: African Americans, African Americans--Enslaved People, African Americans--Enslaved People. Contraband of War, and African Americans--Historically Free and Newly Freed. In order to track which African Americans subject tag the corresponding tag relates to, I added the root or linked subject tag in column A. FTP then exports the corresponding tag in column B. If relevant, latitude and longitude is included in columns C and D respectively. Column F lists the document title and column K includes a description of the document both tags were found in, with a unique FTP identifier listed in column G. Column H lists the year and column J lists the date the document was written, and who authored it can be found in column I.  Column L lists geographic locations relevant to the document, including the location it was sent from is listed. Finally, column M lists the CWRGM document identifier, so the related documents can be found at the website.

<b>Research Question 1:</b> Did Americansâ€™ priorities surrounding Black-white race relations increase or decrease throughout the American Civil War (1861â€“1865) and Reconstruction (May 1865â€“1877) in Mississippi, and if so, how and why?

<b>Approach:</b> One way to begin researching this question is to compare how frequently the letter authors discussed African Americans, Enslaved people, and free and freed people based on the month and year. This will not answer why, but it can point out important time spans for deeper analysis.

```python
# Define a function to convert the date and handle errors
def convert_to_year_month(date_str):
    for fmt in ('%Y-%m-%d', '%Y-%m'):
        try:
            # Try parsing the date with the given format
            date = pd.to_datetime(date_str, format=fmt)
            # Extract the year and month
            return date.strftime('%Y-%m')
        except ValueError:
            continue
    # Return NaN if all formats fail
    return np.nan
```

```python
# Apply the function to the date_column
df['year_month'] = df['Date'].apply(convert_to_year_month)

# Drop rows with NaN in the year_month column
df = df.dropna(subset=['year_month'])

# Filter rows where subjects is 'African Americans'
df_filtered = df[df['Linked Subject Tag'] == 'African Americans']

# Group by year_month and count the occurrences
df_grouped = df_filtered.groupby('year_month').size().reset_index(name='count')

# Plot the timeline using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='count', data=df_grouped, marker='o', color='navy')
plt.xticks(rotation=45)
plt.title('Number of Times "African Americans" was Mentioned by Year and Month')
plt.xlabel('Year-Month')
plt.ylabel('"African Americans" mentions')
plt.tight_layout()

#Display every 2 tick on the x axis so they are readable
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::2])

# Show the plot
plt.show()
```

You'll notice major upticks in May 1874 and June 1875. The May 1874 uptick may have been caused by catastrophic flooding of the Mississippi River that spring, which severely impacted African Americans. The June 1875 uptick is less clear, however, and may point to a fruitful area of further research for scholars. Ultimately, the uptick in references to African Americans from 1874â€“1875 is unsurprising, as white Mississippians drafted, unveiled, and put into effect the "Mississippi Plan," which sought to instititute racial segregation full scale and bring an end Reconstruction and increase Black citizenship rights. 

Do these trends hold true when we look at references to enslaved people and free and freed people specifically?

```python
df['year_month'] = df['Date'].apply(convert_to_year_month)
df = df.dropna(subset=['year_month'])

# Filter rows where subjects is 'African Americans--Enslaved People'
df_filtered_enslaved = df[df['Linked Subject Tag'] == 'African Americans--Enslaved People']
df_grouped_enslaved = df_filtered_enslaved.groupby('year_month').size().reset_index(name='count')
plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='count', data=df_grouped_enslaved, marker='o', color='navy')
plt.xticks(rotation=45)
plt.title('Number of Times "African Americans--Enslaved People" was Mentioned by Year and Month')
plt.xlabel('Year-Month')
plt.ylabel('"African Americans--Enslaved People" mentions')
plt.tight_layout()
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::2])
plt.show()
```

As expected, there is an increase in the number of times enslaved people were mentioned in early 1862 through the spring of 1865. The increase in the spring of 1862 likely corresponds to a number of evolving Union and Confederate military policies, that increasingly entrenched slavery as the central issue of the war. The Confederacy instituted a national draft in April 1862, which allowed drafted men to hire substitutes to service in their place and excused men from military service for medical reasons. That October, the government revised the draft to excuse men enslaved a minimum of twneyt (late fifteen) people. Compounding this, was a serious of Union policies. Enacted in July 1862, the U.S. government passed the Second Confiscation Act, which declared the slaves of those openly participating in the rebellion were free and the Militia Act of 1862, which allowed freed people to perform roles for Union armies. These acts encouraged greater numbers of enslaved people to emancipate themselves, supplied manpower to the Union, and deprived the Confederacy of enslaved
peopleâ€™s labor, which was essential to the Confederate war effort. The highest increase on the chart is in the lead up to and during the passage of the Emancipation Proclamation. Following Union victory at the Battle of Antietam, US President Abraham Lincoln issued the preliminary Emancipation Proclamation in September 1862, which would free all enslaved people in states who seceded from the Union took effect on January 1, 1863. The final wartime uptick occurs during the Confederacy's downfall throughout the spring on 1865, a moment when Americans everywhere were questions what would become of enslaved African Americans. Interestingly, however, there is also a spike in the number of times enslaved people were mentioned in the summer of 1874 that is worthy of further investigation.

```python
df['year_month'] = df['Date'].apply(convert_to_year_month)
df = df.dropna(subset=['year_month'])

# Filter rows where subjects is 'African Americans--Historically Free and Newly Freed'
df_filtered_freed = df[df['Linked Subject Tag'] == 'African Americans--Historically Free and Newly Freed']
df_grouped_freed = df_filtered_freed.groupby('year_month').size().reset_index(name='count')
plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='count', data=df_grouped_freed, marker='o', color='navy')
plt.xticks(rotation=45)
plt.title('Number of Times "African Americans--Historically Free and Newly Freed" was Mentioned by Year and Month')
plt.xlabel('Year-Month')
plt.ylabel('"African Americans--Historically Free and Newly Freed" mentions')
plt.tight_layout()
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::4])
plt.show()
```

Understandably, there are also very few mentions of free and freed people throughout the war. We see a sharp rise, however, when the war ends in the summer of 1865, which likely corresponds to postwar questions about the status of enslaved people. Then there are higher mentions throughout Reconstruction than there were during the war, until 1874â€“1875 when the number of mentions increase dramatically as white Mississippians plotted to roll out segregation and curtail freed people's rights with the adoption of the Mississippi Plan.

These visualizations allow us to track how frequently letter authors were discussing African Americans throughout the war and Reconstruction, but it does not offer any insights into what topics were being discussed in conjunction with the African Americans subject tags. Fortunately, the data includes the other key terms that were discussed alongside these tags. While being mentioned in the same letter does not automatically equate a correlation, tracking the frequency of co-occuring subject tags will point scholars in the direction of key themes.

<b>Research Question 2:</b> How did the transition from slavery to freedom unfold in Mississippi during the war and Reconstruction?

<b>Approach:</b> One way to begin researching this question is to compare which topics were discussed most frequently in connection to the African Americans subject tags. This will not answer why, but it can point out important topics for deeper analysis.

```python
df_filtered = df[df['Linked Subject Tag'] == 'African Americans']
df_grouped = df_filtered.groupby('Cooccuring Subject Title').size().reset_index(name='count')
df_grouped_count=df_grouped.groupby('Cooccuring Subject Title', as_index=False)['count'].sum().sort_values(by='count', ascending=False).head(20)

sns.set(rc={'figure.figsize':(16,9)})

sns.barplot(x='Cooccuring Subject Title', y='count', data=df_grouped_count, color='navy').set(xlabel='Co-occuring Subject Tag', ylabel='Frequency by Document', title='Subjects Most Commonly Discussed in Documents Mentioning "African Americans"');
plt.xticks(rotation=90)
plt.show()
```

The results here do not reveal anything particularly interesting. The three most commonly co-occuring subjects here are Mississippi Governor Adelbert Ames, Secretaries, and the Mississippi Executive Department. Considering that the Ames collection comprisese a significant number of letters in the dataset, and that secretataries signed many of the documents sent to the Mississippi Executive Department, these results are not suprising. The remaining results, however, indicate a high prevalence of subject tags pertaining to the criminal justice system. We will see if this trend continues when looking explicitly at documents referencing enslaved, and free and freed people. 

```python
df_filtered = df[df['Linked Subject Tag'] == 'African Americans--Enslaved People']
df_grouped = df_filtered.groupby('Cooccuring Subject Title').size().reset_index(name='count')
df_grouped_count=df_grouped.groupby('Cooccuring Subject Title', as_index=False)['count'].sum().sort_values(by='count', ascending=False).head(20)

sns.set(rc={'figure.figsize':(16,9)})

sns.barplot(x='Cooccuring Subject Title', y='count', data=df_grouped_count, color='navy').set(xlabel='Co-occuring Subject Tag', ylabel='Frequency by Document', title='Subjects Most Commonly Discussed in Documents Mentioning "Enslaved People"');
plt.xticks(rotation=90)
plt.show()
```

Similar to the previous chart, it is unsurprising that Mississippi Governore John J. Pettus is the most commonly co-occuring subject, as Gov. Pettus was Mississipp's wartime governor. Unlike the previous example, however, subjects pertaining to the criminal justic system are absent in this chart. Rather, key wartime topics (such as military units and procedures) and key roles within slavery (such as enslavers and plantation overseers), are prevelent. 

```python
df_filtered = df[df['Linked Subject Tag'] == 'African Americans--Historically Free and Newly Freed']
df_grouped = df_filtered.groupby('Cooccuring Subject Title').size().reset_index(name='count')
df_grouped_count=df_grouped.groupby('Cooccuring Subject Title', as_index=False)['count'].sum().sort_values(by='count', ascending=False).head(20)

sns.set(rc={'figure.figsize':(16,9)})

sns.barplot(x='Cooccuring Subject Title', y='count', data=df_grouped_count, color='navy').set(xlabel='Co-occuring Subject Tag', ylabel='Frequency by Document', title='Subjects Most Commonly Discussed in Documents Mentioning "Historically Free and Newly Freed People"');
plt.xticks(rotation=90)
plt.show()
```

When looking at the documents containing references to free and freed people,the prevelence of subjects pertaining to the criminal justice system returns. Its absence in documents referencing enslaved people may point a shift in Black-white race relations. Prior to the passage of the Thirteenth Amendment, slavery governed relations between white and Black Mississippians. After slavery ended, however, white southerners turned to the criminal justice system to exert power over freed people by instituting convict leasing programs and laws that funneled Black Americans into the state's penitentiary system and then leased those convicted out to private companies to perform unpaid labor.
